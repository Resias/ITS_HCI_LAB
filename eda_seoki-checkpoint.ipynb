```python
import pandas as pd
import matplotlib.pyplot as plt

# 승차역별 이용건수 집계
station_counts = df["승차역명"].value_counts()

# 5000건 이상인 역만 필터링
station_counts_filtered = station_counts[station_counts >= 5000]

# 막대그래프
plt.figure(figsize=(12,6))
station_counts_filtered.plot(kind="bar", color="skyblue")

plt.title("승차역별 이용 건수 (5000건 이상)")
plt.xlabel("승차역명")
plt.ylabel("이용 건수")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 2호선 역 리스트 (51개) – 이미 따로 df2line 으로 관리하고 있다고 가정
line2_stations = [
    "시청", "을지로입구", "을지로3가", "을지로4가", "동대문역사문화공원", "신당", "상왕십리", "왕십리(성동구청)",
    "한양대", "뚝섬", "성수", "건대입구", "구의", "강변", "잠실나루", "잠실", "잠실새내", "종합운동장",
    "삼성", "선릉", "역삼", "강남", "교대", "서초", "방배", "사당", "낙성대", "서울대입구", "봉천",
    "신림", "신대방", "구로디지털단지", "대림", "신도림", "문래", "영등포구청", "당산", "합정", "홍대입구",
    "신촌", "이대", "아현", "충정로", "용답", "신답", "신설동", "도림천", "양천구청", "신정네거리", "용두"
]

# 1. 2호선 승하차 데이터만 필터링
df_line2 = df[df["승차역명"].isin(line2_stations) | df["하차역명"].isin(line2_stations)]

# 2. OD 매트릭스 생성 (pivot_table)
od_matrix = pd.pivot_table(
    df_line2,
    index="승차역명",   # 행
    columns="하차역명", # 열
    values="가상카드번호",  # 단순 count용 → 아무 컬럼이나 써도 됨
    aggfunc="count",
    fill_value=0
)

# 3. 역 순서를 고정 (51개 순서대로 보기 좋게)
od_matrix = od_matrix.reindex(index=line2_stations, columns=line2_stations, fill_value=0)

# 결과 보기
print(od_matrix.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.heatmap(od_matrix, cmap="Blues", linewidths=.5)
plt.title("서울지하철 2호선 OD 매트릭스 (승차역→하차역)")
plt.xlabel("하차역")
plt.ylabel("승차역")
plt.show()
```
